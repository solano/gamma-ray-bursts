### Mean density

In the case of fast cooling ie $\gamma_m>\gamma_c$, we have two possible situations:
- $\gamma_m>\gamma_e$ in which case the integral must be performed between $t_m$ and $t_{\infty}$.
- $\gamma_e>\gamma_m$ in which case the integral must be performed between 0 and $t_{\infty}$.

#### The case $\gamma_m>\gamma_e$
In the $1^{\text {st }}$ case, the integral reads: $\overline{n\left(\gamma_e\right)}=\int_{t_m}^{t_{\infty}} \gamma_0\left(\gamma_e, t\right)^{-p} \frac{\gamma_0^2}{\gamma_e^2} e^{-\frac{t}{t_{dyn}^{\prime}}} d t$
Defining $x_m=e^{-\frac{t_m}{t_{d y n}^{\prime}}}$ and $x_{\infty}=e^{-\frac{t_{\infty}}{t_{d y n}}}$ and using : $\gamma_0\left(x, \gamma_e\right)=\frac{\gamma_c}{\left(1+\frac{\gamma_c}{\gamma_e}\right) x-1}$ with $x=e^{-\frac{t}{t_{d y n}^{\prime}}}$, we eventually get :
$$
\overline{n\left(\gamma_e\right)}=-\gamma_e^{-2} \int_{x_m}^{x_{\infty}}\left(\frac{\gamma_c}{\left(1+\frac{\gamma_c}{\gamma_e}\right) x-1}\right)^{-p+2} x d x
$$
Defining $u_m=\left(1+\frac{\gamma_c}{\gamma_e}\right) x_m-1$ and $u_{\infty}=\left(1+\frac{\gamma_c}{\gamma_e}\right) x_{\infty}-1$, we finally have:
$$
\begin{gathered}
\overline{n\left(\gamma_e\right)}=\gamma_e^{-2} \gamma_c^{-p+2} \int_{u_{\infty}}^{u_m} \frac{\frac{1}{u}^{-p+2}(u+1)}{\left(1+\frac{\gamma_c}{\gamma_e}\right)^2} d u \\
\overline{n\left(\gamma_e\right)}=\frac{\gamma_e^{-2}}{\left(1+\frac{\gamma_c}{\gamma_e}\right)^2} \gamma_c^{-p+2} \int_{u_{\infty}}^{u_m} u^{p-1}+u^{p-2} d u 
\end{gathered}
$$

Using $t_m=t_{d y n}^{\prime}\left(\ln \left(1+\frac{\gamma_c}{\gamma_e}\right)-\ln \left(1+\frac{\gamma_c}{\gamma_m}\right)\right)$, we retrieve $x_m=e^{-\frac{t_m}{t_{d y n}^{\prime}}}=\frac{1+\frac{\gamma_c}{\gamma_m}}{1+\frac{\gamma_c}{\gamma_e}}$ and then $u_m=\frac{\gamma_c}{\gamma_m}$. Likewise, from $t_{\infty}=t_{d y n}^{\prime} \ln \left(1+\frac{\gamma_c}{\gamma_e}\right)$, we get $x_{\infty}=e^{-\frac{t_{\infty}}{t_{d y n}}}=\frac{1}{1+\frac{\gamma_c}{\gamma_\theta}}$ then $u_{\infty}=0 $


Finally, we rewrite :
$$
\begin{aligned}
& \overline{n\left(\gamma_e\right)}=\frac{\gamma_e^{-2}}{\left(1+\frac{\gamma_c}{\gamma_e}\right)^2} \gamma_c^{-p+2}\left[\frac{u^p}{p}+\frac{u^{p-1}}{p-1}\right]_0^{u_m} \propto \frac{\gamma_e^{-2}}{\left(1+\frac{\gamma_c}{\gamma_e}\right)^2} \approx \gamma_e^{-2} \text { for } \gamma_e \gg \gamma_c
\end{aligned}
$$
We get $\overline{n\left(\gamma_e\right)} \propto \gamma_e^{-2}$ if $\gamma_e<\gamma_m$.

#### The case $\gamma_e>\gamma_m$
In the second case, the first boundary is 0 . The integral reads:
$$
\begin{gathered}
\overline{n\left(\gamma_e\right)}=\int_0^{t_{\infty}} \gamma_0\left(\gamma_e, t\right)^{-p} \frac{\gamma_0^2}{\gamma_e^2} e^{-\frac{t}{t^{\prime}} d y n} d t \text { yielding: } \\
\overline{n\left(\gamma_e\right)}=-\gamma_e^{-2} \int_{x_m=1}^{x_{\infty}}\left(\frac{\gamma_c}{\left(1+\frac{\gamma_c}{\gamma_e}\right) x-1}\right)^{-p+2} x d x
\end{gathered}
$$
Using the same trick, we find that $u_m=\frac{\gamma_c}{\gamma_e}$ and $u_{\infty}=0$.
$\overline{n\left(\gamma_e\right)}=\frac{\gamma_e^{-2}}{\left(1+\frac{\gamma_c}{\gamma_e}\right)^2} \gamma_c^{-p+2}\left[\frac{u^p}{p}+\frac{u^{p-1}}{p-1}\right]_0^{u_m=\frac{\gamma_c}{\gamma_e}} \approx \gamma_e^{-2} \gamma_c^{-p+2} \frac{\left(\frac{\gamma_c}{\gamma_e}\right)^{p-1}}{p} \propto \gamma_e^{-(p+1)}$ Because we only take the dominant term in $\gamma_e$

## Mean energy


Synchrotron theory gives that : 

$$\left\{\begin{array}{c}P(\gamma)=K_1 B^2 \gamma^2 \\ \nu_{s y n}(\gamma)=K_2 B \gamma^2 \\ P_\nu(\gamma)=\frac{P(\gamma)}{\nu_{s y n}(\gamma)} \Phi\left(\frac{\nu}{\nu_{s y n}(\gamma)}\right)\end{array}\right.$$

where

$$\Phi(x)=\left\{\begin{array}{c}\frac{4}{3} x^{\frac{1}{3}}, x<1 \\ 0, x \geq 1\end{array}\right.$$

In fact, above the synchrotron frequency, there is an exponential decay, which is here modeled with a hard cut-off.

The typical timescale of the expansion is $t_{d y n}^{\prime}=\frac{R}{\Gamma_r \beta c}$ with $R$ the average radius of the expansion, $\Gamma_r$ the Lorentz factor to move back to observer frame and $\beta c$ the velocity of the expansion.




We now want to define the following quantities:
- The power spectrum per unit volume $P_\nu=\int_0^{\infty} n(\gamma) P_\nu(\gamma) d \gamma$ at a given frequency $\nu$
- The energy density per unit volume $u_\nu=\int_0^{t_{d y n}^{\prime}} d t^{\prime} \int_0^{\infty} n(\gamma, t) P_\nu(\gamma) d \gamma$ a at a given frequency $\nu$. Assuming the density evolves smoothly during the timescale of expansion $t_{d y n}^{\prime}$, we can use the average density i.e. $u_\nu=t_{d y n}^{\prime} \int_0^{\infty} \bar{n}(\gamma) P_\nu(\gamma) d \gamma$


### Power spectrum 

Power spectrum:
$$
P_\nu=\int_{\gamma_c}^{\infty} n(\gamma) P_\nu(\gamma) d \gamma \propto \int_{\nu_c}^{\infty} n\left(\gamma\left(\nu_s\right)\right) \Phi\left(\frac{\nu}{\nu_s}\right) \frac{d \nu_s}{\sqrt{\nu_s}}
$$
- If $\nu<\nu_c$
$$
P_\nu \propto \nu^{\frac{1}{3}} \underbrace{\int_{\nu_c}^{\infty} n\left(\gamma\left(\nu_s\right)\right) \nu_s^{-\frac{1}{3}} \frac{d \nu_s}{\sqrt{\nu_s}}}_{\text {Constant }} \propto \nu^{\frac{1}{3}}
$$
- If $\nu>\nu_c$
$$
\begin{gathered}
P_\nu \propto \nu^{\frac{1}{3}} \int_\nu^{\infty} n\left(\gamma\left(\nu_s\right)\right) \nu_s^{-\frac{1}{3}} \frac{d \nu_s}{\sqrt{\nu_s}} \\
\int_\nu^{\infty} n\left(\gamma\left(\nu_s\right)\right) \nu_s^{-\frac{1}{3}} \frac{d \nu_s}{\sqrt{\nu_s}} \propto \int_\nu^{\infty} \nu_s^{-\frac{p}{2}} \nu_s^{-\frac{1}{3}} \frac{d \nu_s}{\sqrt{\nu_s}} \propto \nu^{-\frac{p+1}{2}-\frac{1}{3}} \\
\rightarrow P_\nu \propto \nu^{-\frac{p+1}{2}}
\end{gathered}
$$


We now compute the energy density $u_\nu$ :
$$
\begin{gathered}
\int_{\gamma_c}^{\infty} \bar{n}(\gamma) P_\nu(\gamma) d \gamma=\int_{\gamma_c}^{\infty} \frac{4 \bar{n}(\gamma) K_1 B}{3 K_2} \Phi\left(\frac{\nu}{\nu_{s y n}(\gamma)}\right) \frac{d \nu_{s y n}}{2 \sqrt{K_2 B \nu_{s y n}}} \\ \propto \int_{\nu_c}^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \bar{n}\left(\gamma\left(\nu_s\right)\right) \Phi\left(\frac{\nu}{\nu_s}\right)\\
\end{gathered}
$$

We have previously shown that :$$\begin{cases}
\bar{n}(\gamma) \propto \gamma^{-2},&\gamma_c \leq \gamma \leq \gamma_m \\
\bar{n}(\gamma) \propto \gamma^{-(p+1)},&\gamma_m \leq \gamma
\end{cases}$$


- If $\nu<\nu_c$: $\int_{\gamma_c}^{\infty} \bar{n}(\gamma) P_\nu(\gamma) d \gamma \propto \nu^{\frac{1}{3}} \int_{\nu_c}^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \bar{n}\left(\gamma\left(\nu_s\right)\right) \nu_{\mathrm{s}}^{-\frac{1}{3}} $. Therefore : $u_{\nu} \propto \nu^{\frac{1}{3}}$.
- If $\nu_c<\nu<\nu_m$: $\int_{\gamma_c}^{\infty} \bar{n}(\gamma) P_\nu(\gamma) d \gamma \propto \nu^{\frac{1}{3}} \int_\nu^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \bar{n}\left(\gamma\left(\nu_s\right)\right) \nu_{\mathrm{s}}^{-\frac{1}{3}}$


With $\int_\nu^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \bar{n}\left(\gamma\left(\nu_s\right)\right) \nu_s^{-\frac{1}{3}}=\underbrace{\int_\nu^{\nu_m} \frac{d \nu_s}{\sqrt{\nu_s}} \nu_s^{-1} \nu_s^{-\frac{1}{3}}}_{\text {Dominant }}+\int_{\nu_m}^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \nu_s^{-\frac{p}{2}} \nu_s^{-\frac{1}{3}}\approx \int_\nu^{\nu_m} \frac{d \nu_s}{\sqrt{\nu_s}} \nu_s^{-1} \nu_s^{-\frac{1}{3}} \propto \left[\nu_s^{-\frac{1}{2}-\frac{1}{3}}\right]_{\nu}^{\nu_m}$

This finally yields $u_{\nu} \propto \nu^{-\frac{1}{2}}$
- If $\nu>\nu_m$: $\int_{\gamma_c}^{\infty} \bar{n}(\gamma) P_\nu(\gamma) d \gamma \propto \nu^{\frac{1}{3}} \int_\nu^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \bar{n}\left(\gamma\left(\nu_s\right)\right) \nu_s^{-\frac{1}{3}}$

With $\int_\nu^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \bar{n}\left(\gamma\left(\nu_s\right)\right) \nu_s^{-\frac{1}{3}}=\int_{\nu_m}^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \nu_s^{-\frac{p}{2}} \nu_s^{-\frac{1}{3}}$
$\int_{\nu_m}^{\infty} \frac{d \nu_s}{\sqrt{\nu_s}} \nu_s^{-\frac{p}{2}} \nu_s^{-\frac{1}{3}}$

We finally get $u_{\nu} \propto \nu^{-\frac{p}{2}}$


## Shocked matter properties

- $\epsilon_*$ is the average energy dissipated per unit mass in the shock. It is given by : $\epsilon_*=(\Gamma_{int}-1)m_pc^2$  where $\Gamma_{\text {int }}=\frac{1}{2}\left[\left(\frac{\Gamma_1}{\Gamma_2}\right)^{1 / 2}+\left(\frac{\Gamma_2}{\Gamma_1}\right)^{1 / 2}\right]$,
where $\Gamma_{\text {int }}$ is the Lorentz factor for internal motions in the shocked material. 
- $\rho_*$ is the mass density of shocked matter in comoving frame.
- $u_*$ is the internal energy density of shocked matter in comoving frame.
- $B$ is the amplified magnetic field in comoving frame.
- $t_{dyn}^{\prime}$ is the dynamical shell expansion time.
- $e$ is the amount of energy available to dissipate in observer frame when an equal mass has been swept. Concretely:
$e=mc^2(\Gamma_1+\Gamma_2-2\Gamma_r)$ with $m=min(\Gamma_1,\Gamma_2)$
- $\Gamma_f$ is the ultrarelativistic Lorentz factor of resulting shell :
$\Gamma_f=\sqrt{\Gamma_1\Gamma_2\frac{m_1\Gamma_1+m_2\Gamma_2}{m_1\Gamma_2+m_2\Gamma_1}}$
- When an equal mass in both shells has been swept (i.e $m_1=m_2$), $\Gamma_f=\Gamma_r=\sqrt{\Gamma_1\Gamma_2}$.
- $t_{obs}=t+\frac{D-R}{c}$ is the time in the observer frame. $D$ is the order of several Mpc, which is why it will be dropped from now on : $t_{obs}\equiv t-\frac{R}{c}$


We want to evaluate the typical timescale in which the shocked matter cools down by pure expansion $t_{dyn}^{\prime}$ (expansion time).

We start from the thermodynamical identity : $\frac{dU}{dt}=-P\frac{dV}{dt}$

Using perfect gas equation of state : $P=\alpha u$ with $u=\frac{U}{V}$ the density of energy.
This yields : $\frac{dU}{dt}=-\alpha \frac{U}{V} \frac{dV}{dt}$ or equivalently $\frac{dln(U)}{dt}=-\alpha \frac{dln(V)}{dt}$.

The expansion time is the typical timescale of the loss of energy via expansion ie $-\frac{1}{t_{dyn}}\equiv\frac{dln(U)}{dt}$.

With $V\propto R^x$, we have for the right-hand term $-\alpha \frac{dln(V)}{dt}=-\alpha x \frac{dln(R)}{dt}=-\alpha x \frac{v}{R}=$.

Thus, we get in the observer frame : $t_{dyn}=\frac{R}{\beta c}$.
In comoving frame : $t_{dyn}^{\prime}=\frac{R}{\Gamma\beta c}$
