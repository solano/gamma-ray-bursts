```python
import numpy as np
import matplotlib.pyplot as plt

def fun(t, boxsize):
    if 0 <= t and t < boxsize:
        return 2/boxsize * (1 - t/boxsize)
    return 0

fun = np.vectorize(fun, excluded='boxsize')

def fun2(t, boxsize):
    return (0<=t).astype(int)*(t<boxsize).astype(int)*(2/boxsize)*(1 - t/boxsize)

t = np.linspace(-0.1, 1, 500)
plt.plot(t, fun(t, 0.4))
plt.plot(t, fun2(t, 0.4))
```
